```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.linspace(5, 100, 100)
y = np.linspace(10, 1000, 100)

plt.plot(x, y)
```

```python
font = {'family': 'serif',
        'color':  'white',
        'weight': 'normal',
        'size': 16,
        }
plt.xlabel('Current', fontdict=font)
plt.ylabel('Voltage', fontdict=font)
plt.title('Ohm''s law')
  
plt.plot(x, y, 'r.')
plt.show()
```

```python
x = np.linspace(1, 10, 100)
y = np.log(x)

plt.figure(1)

plt.subplot(1, 2, 1)
plt.title('y = log(x)')
plt.plot(x, y)

plt.subplot(1,2, 2)
plt.title(' y = x**2')
plt.plot(x, x**2)
```
